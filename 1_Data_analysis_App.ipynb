```python
%%writefile app.py
# Import libraries
import streamlit as st
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

st.title("Data Analysis App")
st.subheader("This App will show Data Analysis Basic Plots")

# Select the dataset from seaborn
select_df = ['tips', 'titanic', 'iris']
df_name = st.selectbox('Select Dataset', select_df)

if df_name == 'tips':
  df = sns.load_dataset('tips')

elif df_name == 'titanic':
  df = sns.load_dataset('titanic')

elif df_name == 'iris':
  df = sns.load_dataset('iris')

# Upload you're own dataset
df_upload = st.file_uploader('Upload Dataset', type=['csv', 'xlsx'])

if df_upload is not None:

    df = pd.read_csv(df_upload)

# Display the fisrt values
st.write(df)

# Display the shape
st.write(df.shape)

# Display the number of rows and columns
st.write(df.shape[0], 'rows', df.shape[1], 'columns')

# Display the data type
st.write(df.dtypes)

# Display the missing values
st.write(df.isnull().sum().sort_values(ascending=False))

# Display the summary statistics
st.write(df.describe())

# Select the specific columns from dataset for plots
X_col = st.selectbox('Select the independent columns', df.columns)
y_col = st.selectbox('Select the dependent column', df.columns)
plot_type = st.selectbox('Select Plots Type', ['line', 'bar', 'scatter', 'hist', 'box', 'violin'])

if plot_type == 'line':
 st.line_chart(df[[X_col, y_col]])

elif plot_type == 'bar':
 st.bar_chart(df[[X_col, y_col]])

elif plot_type == 'scatter':
 st.scatter_chart(df[[X_col, y_col]])

elif plot_type == 'hist':
 st.hist_chart(df[[X_col, y_col]])

elif plot_type == 'box':
 st.box_chart(df[[X_col, y_col]])

elif plot_type == 'violin':
 st.violin_chart(df[[X_col, y_col]])

```
