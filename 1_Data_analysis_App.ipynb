```python
%%writefile app.py
# Import libraries
import streamlit as st
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

st.title("Data Analysis App")
st.subheader("This App will show Data Analysis Basic Plots")

# Select the dataset from seaborn
select_df = ['tips', 'titanic', 'iris']
df_name = st.selectbox('Select Dataset', select_df)

if df_name == 'tips':
  df = sns.load_dataset('tips')

elif df_name == 'titanic':
  df = sns.load_dataset('titanic')

elif df_name == 'iris':
  df = sns.load_dataset('iris')

# Upload you're own dataset
df_upload = st.file_uploader('Upload Dataset', type=['csv', 'xlsx'])

if df_upload is not None:

    df = pd.read_csv(df_upload)

# Display the fisrt values
st.write(df)

# Display the shape
st.write(df.shape)

# Display the number of rows and columns
st.write(df.shape[0], 'rows', df.shape[1], 'columns')

# Display the data type
st.write(df.dtypes)

# Display the missing values
st.write(df.isnull().sum().sort_values(ascending=False))

# Display the summary statistics
st.write(df.describe())

# Plot the columns by PairPlor

st.subheader('PairPlot')

hue_col = st.selectbox('Select the hue column', df.columns)
st.pyplot(sns.pairplot(df, hue=hue_col))

import streamlit as st
import plotly.express as px

# Correlation matrix
corr = df.corr(numeric_only=True)

# Create heatmap
fig = px.imshow(
    corr,
    text_auto=True,   # show values
    aspect="auto",
    color_continuous_scale="Viridis"
)

# Show in Streamlit
st.plotly_chart(fig)
```
